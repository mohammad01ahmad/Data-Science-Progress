```python
df['Timestamp'] = pd.to_datetime(df['Timestamp'])
```

```python
df['Hour'] = df['Timestamp'].dt.hour
df.head(3)
```

```python
df['Month'] = df['Timestamp'].dt.month
df['DayOfWeek'] = df['Timestamp'].dt.dayofweek  # 0 = Monday, 6 = Sunday
df['Year'] = df['Timestamp'].dt.year
df.head()
```

```python
df.columns
```

```python
X = df[['Daily Time Spent on Site', 'Age', 'Area Income', 'Daily Internet Usage', 'Male', 'Hour', 'Day', 'Month', 'DayOfWeek']]
y= df['Clicked on Ad']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
Log = LogisticRegression()
```

```python
Log.fit(X_train, y_train)
```

```python
lol = Log.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, lol))
```
