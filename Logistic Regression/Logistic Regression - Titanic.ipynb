```python
def impute_age(cols):
    Age = cols[0]
    Pclass = cols[1]

    if pd.isnull(Age): 

        if Pclass == 1:
            return 37
        if Pclass == 2:
            return 29
        if Pclass == 3:
            return 24
            
    return Age
```

```python
df['Age'] = df[['Age', 'Pclass']].apply(impute_age, axis=1)
```

```python
df.info()
```

```python
df.drop('Cabin', axis=1)
```

```python
sex = pd.get_dummies(df['Sex'], drop_first=True).astype(int)
sex.head()
```

```python
embark = pd.get_dummies(df['Embarked'], drop_first=True).astype(int)
embark.head()
```

```python
df = pd.concat([df, sex, embark], axis=1)
df.head(2)
```

```python
 df.drop(['Name', 'Sex', 'Embarked', 'Ticket'], axis=1, inplace=True)
```

```python
df.head()
```

```python
df.drop('Cabin', axis=1, inplace=True)
df.head()
```

```python
df.drop('PassengerId', axis=1, inplace=True)
df.head()
```

```python
df.columns
```

```python
X = df[['Pclass', 'Age', 'SibSp', 'Parch', 'Fare', 'male', 'Q', 'S']]
y = df['Survived']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
Prediction = lr.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, Prediction))
```

```python
from sklearn.metrics import confusion_matrix
```

```python
print(confusion_matrix(y_test, Prediction))
```
