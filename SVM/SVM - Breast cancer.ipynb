```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
```

```python
cancer.keys()
```

```python
df = pd.DataFrame(cancer['data'], columns=cancer['feature_names'])
df.head()
```

```python
df.info()
```

```python
cancer['target_names']
```

```python
test = df[['mean radius', 'mean texture', 'mean perimeter', 'mean area', 'mean smoothness', 'mean compactness',
'mean concavity','mean concave points', 'mean symmetry', 'mean fractal dimension']]
```

```python
sns.pairplot(data=test)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df
y = cancer['target']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.svm import SVC
```

```python
model = SVC()
```

```python
model.fit(X_test, y_test)
```

```python
pred = model.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, pred))
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {'C':[0.1,1,10,10,1000], 'gamma':[1, 0.1, 0.01, 0.001, 0.0001]}
```

```python
grid = GridSearchCV(SVC(), param_grid, verbose=3)
```

```python
grid.fit(X_train, y_train)
```

```python
grid_predictions = grid.predict(X_test)
```

```python
print(classification_report(y_test, grid_predictions))
```
