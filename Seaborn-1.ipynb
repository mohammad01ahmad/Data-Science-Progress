```python
import seaborn as sns
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.histplot(data=tips, x='total_bill')    # your histogram
plt.xlabel('Total bill ($)')
plt.ylabel('Count')
plt.title('Distribution of Total Bills')
plt.show()
```

```python
sns.histplot(tips['total_bill'], bins=40)
plt.show()
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips)
plt.show()
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips, kind='hex')
plt.show()
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips, kind='reg')
plt.show()
```

```python
sns.pairplot(tips)
plt.show()
```

```python
sns.pairplot(tips, hue='sex')
plt.show()
```

```python
sns.rugplot(tips['total_bill'])
plt.show()
```

```python
#Matrix Plotting
sns.barplot(x='sex',y='total_bill', data=tips)
plt.show()
```

```python
sns.countplot(x='sex', data=tips)
plt.show()
```

```python
sns.boxplot(x='day', y='total_bill', data=tips)
plt.show()
```

```python
sns.boxplot(x='day', y='total_bill', data=tips, hue='smoker')
plt.show()
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='smoker', split=True)
plt.show()
```

```python
sns.stripplot(x='day', y='total_bill', data=tips)
plt.show()
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, jitter= True, hue='sex')
plt.show()
```

```python
#Matrix Plotting
flights = sns.load_dataset('flights')
flights.head()
```

```python
tips.head()
```

```python
tc = tips.corr(numeric_only=True)
```

```python
sns.heatmap(tc, annot=True)
plt.show()
```

```python
fp = flights.pivot_table(index='month', columns='year', values='passengers', observed=False)
```

```python
sns.heatmap(fp, linecolor='white', linewidths='1')
plt.show()
```

```python
sns.clustermap(fp)
plt.show()
```

```python
#Regression Plots
sns.lmplot(x='total_bill', y='tip',data=tips, col='day', hue='sex', aspect=0.6)
plt.show()
```
