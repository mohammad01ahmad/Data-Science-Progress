```python
import pandas as pd
import numpy as np 
import matplotlib as mt
```

```python
df = pd.read_csv('zomato.csv')
df
```

```python
df['restaurant name'] = df['restaurant name'].replace('#',"")
df['restaurant name'] = df['restaurant name'].str.strip("'#@")
df = df.drop(columns = "Unnamed: 0.1")
df = df.drop(columns = "Unnamed: 0")
```

```python
df.dropna(inplace=True)
df.info()
```

```python
import seaborn as sns
sns.histplot(df['avg cost (two people)'])
```

```python
sns.histplot(df['rate (out of 5)'])
```

```python
sns.boxplot(x=df['avg cost (two people)'])
```

```python
df['avg cost (two people)'] = np.log(df['avg cost (two people)'] + 1)
sns.histplot(df['avg cost (two people)'])
```
