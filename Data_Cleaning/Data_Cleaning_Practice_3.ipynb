```python
df['rating'] = df['rating'].replace('TV-MA', 'PG-17')
df['rating'] = df['rating'].replace('TV-G', 'General Audience')
df['rating'] = df['rating'].replace('NR', 'Not Rated')
df['rating'] = df['rating'].replace('UR', 'Not Rated')
df['rating'] = df['rating'].replace('G', 'General Audience')
df['rating'] = df['rating'].replace('TV-Y7-FV', 'TV-Y7')
df['rating'] = df['rating'].replace('NC-17', 'PG-17')
df['rating'] = df['rating'].replace('74 min', '')
df['rating'] = df['rating'].replace('84 min', '')
df['rating'] = df['rating'].replace('66 min', '')
df
```
